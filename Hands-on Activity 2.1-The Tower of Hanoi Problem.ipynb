### Hands-on Activity 2.1: The Tower of Hanoi Problem

```python
A = [4, 3, 2, 1]
B = []
C = []
moves = 0

def show():
    print(f"A: {A}  B: {B}  C: {C}")

print("START: Move ALL disks from A to B")
print("Stage 1: A to B")

show()

def hanoi(n, source, dest, aux):
    global moves
    if n > 0:
        hanoi(n-1, source, aux, dest)
        
        if source == 'A': disk = A.pop()
        elif source == 'B': disk = B.pop()
        else: disk = C.pop()
        
        if dest == 'A': A.append(disk)
        elif dest == 'B': B.append(disk)
        else: C.append(disk)
        
        moves += 1
        print(f"Move {moves}: Disk {disk} {source} to {dest}")
        show()
        
        hanoi(n-1, aux, dest, source)

hanoi(4, 'A', 'B', 'C')
print(f"Stage 1 DONE: All disks moved from A to B")

print("Stage 2: B to C")
hanoi(4, 'B', 'C', 'A')
print(f"Stage 2 DONE: All disks moved from B to C")

print(f"TOTAL MOVES: {moves}")
```

Explanation: 

The Tower of Hanoi solution works by breaking a big problem into smaller, repeating steps. I Think of it like this: to move a stack of disks, you first move all the smaller disks out of the way to a aux rod, then move the biggest disk to where you want it, and finally move the smaller disks back on top. This same three-step process repeats for each smaller pile until you're just moving one disk at a time. The function  keeps calling itself with fewer disks, and because of how it's set up, it automatically follows the rules only moving one disk and never putting a big disk on a small one.
